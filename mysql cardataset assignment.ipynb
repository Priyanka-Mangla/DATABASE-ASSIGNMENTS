4 . Dump all the data into car table 

```python
import csv
with open('cardataset.data','r')as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = i[0].split(',')
        query = "insert into cardataset.car values('{}','{}','{}','{}','{}','{}','{}')".format(str(row_data[0]),str(row_data[1]),str(row_data[2]),str(row_data[3]),str(row_data[4]),str(row_data[5]),str(row_data[6]))
        cursor.execute(query)
mydb.commit()
```

6 . try to group all the data with COL1 AND count occurences of each and every record based on col1 value

7 . Try to filter a record where col 3 value will be 4 .

```python
query = 'select * from cardataset.car where doors = 4'
cursor.execute(query)
cursor.fetchall()
```

8.  Try to update a col 3 value with 8 wherever you have value equal to 2 

```python
query1 = "update cardataset.car set doors = '8' where doors = '2'"
cursor.execute(query1)
mydb.commit()
```
